```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
f = cv.imread(r"D:\Academics\Semester 4\Machine Vision\Assignments\Assignment 1 Photos\emma_gray.jpg")
assert f is not None

t1 = np.linspace(0, 50, 51)
t2 = np.linspace(100, 255, 100)
t3 = np.linspace(150, 255, 105)
t = np.concatenate((t1, t2, t3), axis = 0).astype(np.uint8)

g = cv.LUT(f,t)
plt.figure(figsize=(18,6))

plt.subplot(1,3,1)
plt.plot(t)
plt.grid()
plt.title('Intensity Transformation')

plt.subplot(1,3,2)
plt.imshow(cv.cvtColor(f, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Original')

plt.subplot(1,3,3)
plt.imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Intensity Transformed')
plt.show()
```
