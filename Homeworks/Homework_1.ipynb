```python
import matplotlib.pyplot as plt

m,c = 2, -4
N = 10
x = np.linspace(0, N-1, N).reshape(N,1)
sigma = 10
y = m*x + c + np.random.normal(0, sigma, (N,1))

plt.plot(x, y)
plt.show()
```

```python
X = np.append(np.ones((N,1)), x, axis = 1)
print(X)
```

```python
value = np.linalg.inv(X.T @ X) @ X.T @ y
print(value)
```
